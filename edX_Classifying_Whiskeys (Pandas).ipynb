```python
import numpy as np
import pandas as pd
whisky = pd.read_csv("whiskies.txt")
#Reading a text file and adding it as a column/ variable to a data
whisky["Region"] = pd.read_csv("regions.txt")
whisky.tail()
```

```python
whisky.iloc[0:10]
```

```python
whisky.iloc[5:10, 0:5]
```

```python
whisky.columns
```

```python
flavors = whisky.iloc[:, 2:14]
flavors
```

```python
#Pearson correlation
corr_flavors = pd.DataFrame.corr(flavors)
corr_flavors
```

```python
import matplotlib.pyplot as plt
#to get a large figure
plt.figure(figsize=(10,10))
#use the pseudo color
plt.pcolor(corr_flavors)
#insert a color bar
plt.colorbar()
plt.savefig("corr_flavors.pdf")
#"Heavy body" is strongly associated with "Smokiness"
```

```python
corr_whisky = pd.DataFrame.corr(flavors.transpose())
plt.figure(figsize=(10,10))
plt.pcolor(corr_whisky)
#to trim axis to end at the last value (no whitespace)
plt.axis("tight")
plt.colorbar()
plt.savefig("corr_whisky.pdf")
```

```python
from sklearn.cluster.bicluster import SpectralCoclustering
#Call the method and set n_clusters = no. of regions the whiskys are coming from, which is 6
model = SpectralCoclustering(n_clusters = 6, random_state = 0)
#Fit the "model" to the "corr_whisky" correlation matrix
model.fit(corr_whisky)
model.rows_
```

```python
np.sum(model.rows_, axis = 1)
```

```python
np.sum(model.rows_, axis = 0)
```

```python
model.row_labels_
```

```python
#Extract group labels from model, append to "whisky" data, specify the index explicitly
whisky['Group'] = pd.Series(model.row_labels_, index = whisky.index)
#Reorder the rows in increasing order by group labels
whisky = whisky.iloc[np.argsort(model.row_labels_)]
#Reset the index of the "whisky" dataframe
whisky = whisky.reset_index(drop=True)
#Re-calculate the correlation matrix
correlations = pd.DataFrame.corr(whisky.iloc[:, 2:14].transpose())
#Turn the correlation matrix dataframe into a numpy array
correlations = np.array(correlations)


#Plot the 2 correlation matrics ("corr_whisky" vs. "correlations" - which is the re-ordered version) for comparison
plt.figure(figsize = (14,7))
plt.subplot(121)
plt.pcolor(corr_whisky, cmap="jet")
plt.title("Original")
plt.axis("tight")
plt.subplot(122)
plt.pcolor(correlations, cmap="jet")
plt.title("Rearranged")
plt.axis("tight")
plt.savefig("correlations.pdf")
#We will expect whiskys in the same cluster to be similar in terms of their smokiness/ honey flavor/ etc.
```
